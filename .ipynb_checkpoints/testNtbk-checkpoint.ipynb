```python
import skimage
from skimage import io
import numpy as np
import matplotlib.pyplot as plt
import PIL
%matplotlib inline
skimage.__version__
```

```python
image_test = np.asarray(PIL.Image.open('./res/strzalyTest.png').convert('L'))
plt.imshow(image_test)
print(type(image_test))
print(image_test.shape)
```

```python
from skimage import filters
# coins = skimage.data.coins()
image_filtered = filters.median(image_test, selem=np.ones((5, 5)))

f, (ax0, ax1) = plt.subplots(1, 2, figsize=(15,5))
ax0.imshow(image_test)
ax1.imshow(image_filtered)
```

```python
edges = skimage.feature.canny(image_filtered, sigma=2)

f, (ax0, ax1, ax2) = plt.subplots(1, 3, figsize=(15,5))
ax0.imshow(skimage.feature.canny(image_filtered, sigma=3))
ax1.imshow(skimage.feature.canny(image_filtered, sigma=2))
ax2.
```
